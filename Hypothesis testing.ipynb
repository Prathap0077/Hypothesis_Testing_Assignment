#  Hypothesis Testing Assignment 

###  State the Hypotheses

```python
from math import sqrt
import scipy.stats as stats
```

```python
sample_mean = 3050     
n = 25                     
X_mean = 600               
sigma_X = 25                
sigma_W = 5 * sigma_X       
mu = 1000 + 5 * X_mean
print("Null Hypothesis (H0): The average weekly operating cost = Rs.", mu)
print("Alternative Hypothesis (H1): The average weekly operating cost > Rs.", mu)
```

###  Calculate the Test Statistic

```python
t_stat = (sample_mean - mu) / (sigma_W / sqrt(n))
print(f"Test Statistic (Z) = {t_stat:.2f}")
```

###  Determine the Critical Value

```python
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha)
print(f"Critical Value (Zα) = {critical_value:.2f}")
```

### Make a Decision

```python
if t_stat > critical_value:
    decision = "Reject the Null Hypothesis"
else:
    decision = "Fail to Reject the Null Hypothesis"
print("Decision:", decision)
```

### Conclusion

```python

if t_stat > critical_value:
    conclusion = ("There is sufficient evidence at the 5% significance level "
                  "to conclude that the weekly operating costs have increased "
                  "beyond the theoretical model prediction.")
else:
    conclusion = ("There is insufficient evidence to support the claim that "
                  "the weekly operating costs are higher than predicted by the model.")

print("Conclusion:", conclusion)
```

###  Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-4, 6, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(8,4))
plt.plot(x, y, label='Standard Normal Curve')
plt.fill_between(x, 0, y, where=(x >= critical_value), color='orange', alpha=0.4, label='Rejection Region (α=0.05)')
plt.axvline(t_stat, color='red', linestyle='--', label=f'Test Statistic (Z={t_stat:.2f})')
plt.title('Hypothesis Test for Weekly Operating Cost')
plt.xlabel('Z value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```
